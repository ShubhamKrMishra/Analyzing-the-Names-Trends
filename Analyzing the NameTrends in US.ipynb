```python
years = []
for year in range(1880, 2023):
    years.append(pd.read_csv(f'yob{year}.txt', names = ['Name', 'Gender', 'Babies']))
    years[-1]['Year'] = year
```

```python
total_births = names_1.pivot_table('Babies', index = 'Year', columns = 'Gender', aggfunc =sum)
total_births
```

```python
total_births.plot(title = 'Total births by sex and year')
```

```python
def add_prop(some_group): 
    some_group['Prop'] = some_group.Babies / some_group.Babies.sum()
    return some_group

names = names_1.groupby(['Year', 'Gender']).apply(add_prop)

names
```

```python
def get_top1000(group):
    return group.sort_values(by = 'Babies', ascending = False)[:1000]


grouped = names.groupby(['Year', 'Gender'])

top1000 = grouped.apply(get_top1000)

top1000.reset_index(inplace = True, drop = True)

top1000
```

```python
boys = top1000[top1000.Gender == 'M']
girls = top1000[top1000.Gender == 'F']

total_births = top1000.pivot_table('Babies', index = 'Year', columns = 'Name', aggfunc = sum)

total_births.T
```

```python
subset = total_births [['John', 'Noah', 'Aanya', 'Aarav', 'Emma', 'Mary']]

subset.plot(subplots = True, figsize = (12, 10), grid = False, 
           title = 'Few boys and Girl Names Over Time')
```
